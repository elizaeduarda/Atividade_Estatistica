Programa em Phython que, a partir de um rol de dados, construa a tabela de distribuição de frequência e calcule a média, mediana, moda, amplitude total, variância amostral e desvio padrão.

```python
import math
from collections import Counter
```

*Entrada dos dados*

```python
entrada = input("Digite os dados separados por espaço: ")
dados = list(map(float, entrada.split()))
```

*Organizando os dados*



```python
dados.sort()
n = len(dados)

if n == 0:
    raise SystemExit("Nenhum dado informado.")
```

*Definição de classes*

```python
k = round(math.sqrt(n))
amplitude_total = max(dados) - min(dados)
h = math.ceil(amplitude_total / k)
```

*Tabela de distribuições*

```python
classes = []
fi = []
xm = []
xm_fi = []
fa = []
fr = []

limite_inferior = min(dados)
acumulado = 0

for i in range(k):
    limite_superior = limite_inferior + h

    classe = (limite_inferior, limite_superior)
    freq = sum(1 for x in dados if limite_inferior <= x < limite_superior)

    ponto_medio = (limite_inferior + limite_superior) / 2

    acumulado += freq

    classes.append(classe)
    fi.append(freq)
    xm.append(ponto_medio)
    xm_fi.append(ponto_medio * freq)
    fa.append(acumulado)
    fr.append((freq / n) * 100)

    limite_inferior = limite_superior
```

*Medidas Estatísticas*
- Dados Brutos

```python
media = sum(dados) / n

# Mediana
if n % 2 == 0:
    mediana = (dados[n//2 - 1] + dados[n//2]) / 2
else:
    mediana = dados[n//2]

# Moda
contador = Counter(dados)
maior_freq = max(contador.values())
moda = [num for num, freq in contador.items() if freq == maior_freq]

# Variância e Desvio padrão (brutos)
variancia_bruta = sum((x - media)**2 for x in dados) / (n - 1)
desvio_padrao_bruto = math.sqrt(variancia_bruta)
```

- Dados Agrupados

```python
x_barra = sum(xm_fi) / n

desvio_abs = [abs(x - x_barra) for x in xm]
desvio_quad = [(x - x_barra) ** 2 for x in xm]

variancia_agrupada = sum(fi[i] * desvio_quad[i] for i in range(k)) / (n - 1)
desvio_padrao_agrupado = math.sqrt(variancia_agrupada)
```

*Resultados*

```python
print("\n================ TABELA DE DISTRIBUIÇÃO ================\n")
print("Classe\t\tfi\txm\txm*fi\tfa\tfr(%)\t|xm-x̄|\t(xm-x̄)^2")

for i in range(k):
    print(f"{classes[i]}\t{fi[i]}\t{xm[i]:.2f}\t{xm_fi[i]:.2f}\t"
          f"{fa[i]}\t{fr[i]:.2f}\t{desvio_abs[i]:.2f}\t{desvio_quad[i]:.2f}")

print("\n================ RESULTADOS (DADOS BRUTOS) ================\n")
print(f"Média: {media:.2f}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Amplitude Total: {amplitude_total:.2f}")
print(f"Variância Amostral: {variancia_bruta:.2f}")
print(f"Desvio Padrão Amostral: {desvio_padrao_bruto:.2f}")

print("\n================ RESULTADOS (DADOS AGRUPADOS) ================\n")
print(f"Média Agrupada (x̄): {x_barra:.2f}")
print(f"Variância Agrupada: {variancia_agrupada:.2f}")
print(f"Desvio Padrão Agrupado: {desvio_padrao_agrupado:.2f}")
```
